```python
import matplotlib.pyplot as plt
import numpy as np

def fit_and_visualize_mixed(square_size, rect_width, rect_height):
    """
    Calculate and visualize how many rectangles fit into a square, using both orientations.
    
    Args:
    - square_size (int): Size of the square (square_size x square_size).
    - rect_width (int): Width of the rectangle.
    - rect_height (int): Height of the rectangle.
    """
    
    grid = np.zeros((square_size, square_size))  # 2D array to track placement (0 = empty, 1 = filled)
    
    rects = []  # List to store placed rectangles
    count = 0  # Number of rectangles placed

    # Try filling the grid row-wise, alternating orientations if necessary
    y = 0
    while y < square_size:
        x = 0
        while x < square_size:
            if grid[y, x] == 0:  # Check if space is available
                
                # Try placing the rectangle in normal orientation (width x height)
                if x + rect_width <= square_size and y + rect_height <= square_size:
                    rects.append((x, y, rect_width, rect_height))
                    grid[y:y+rect_height, x:x+rect_width] = 1
                    count += 1
                    x += rect_width  # Move to the next column
                
                # If normal orientation doesn't fit, try rotated (height x width)
                elif x + rect_height <= square_size and y + rect_width <= square_size:
                    rects.append((x, y, rect_height, rect_width))
                    grid[y:y+rect_width, x:x+rect_height] = 1
                    count += 1
                    x += rect_height  # Move to the next column
                
                else:
                    x += 1  # Move right if no rectangle fits
            else:
                x += 1  # Move right if space is occupied
        
        y += 1  # Move to the next row
    
    # Visualization using Matplotlib
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(0, square_size)
    ax.set_ylim(0, square_size)
    ax.set_xticks(np.arange(0, square_size + 1, rect_width))
    ax.set_yticks(np.arange(0, square_size + 1, rect_height))
    ax.grid(True)

    # Draw rectangles
    for x, y, w, h in rects:
        rect = plt.Rectangle((x, y), w, h, edgecolor='blue', facecolor='lightblue', lw=2)
        ax.add_patch(rect)

    ax.set_title(f"Mixed Orientation Packing: {count} Rectangles in {square_size}x{square_size}")
    plt.gca().invert_yaxis()  # Invert y-axis for correct display
    plt.show()

# Example Usage
square_size = 200  # 200x200 square
rect_width, rect_height = 70, 40  # Rectangle size
fit_and_visualize_mixed(square_size, rect_width, rect_height)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def fit_cells_in_panel(panel_size, cell_width, cell_height, keep_out=None, show_plot=True):
    """
    Calculates and visualizes how many cells fit into a panel, avoiding a keep-out area.

    Args:
    - panel_size (int): Size of the panel (panel_size x panel_size).
    - cell_width (int): Width of the cell.
    - cell_height (int): Height of the cell.
    - keep_out (tuple): (x, y, width, height) defining the keep-out area. Default is None.
    - show_plot (bool): Whether to display the plot (default is True).

    Returns:
    - (int, tuple): Number of cells that fit in the panel and panel details (panel size, cell width, cell height).
    """
    
    grid = np.zeros((panel_size, panel_size))  # 2D array to track placement (0 = empty, 1 = filled)
    
    # Mark the keep-out area in the grid if provided
    if keep_out:
        kx, ky, kw, kh = keep_out
        grid[ky:ky+kh, kx:kx+kw] = -1  # Mark as unavailable

    cells = []  # List to store placed cells
    count = 0  # Number of cells placed

    # Try filling the grid row-wise, alternating orientations if necessary
    y = 0
    while y < panel_size:
        x = 0
        while x < panel_size:
            if grid[y, x] == 0:  # Check if space is available
                
                # Try placing the cell in normal orientation (width x height)
                if x + cell_width <= panel_size and y + cell_height <= panel_size:
                    if np.all(grid[y:y+cell_height, x:x+cell_width] == 0):  # Ensure no overlap with keep-out area
                        cells.append((x, y, cell_width, cell_height))
                        grid[y:y+cell_height, x:x+cell_width] = 1
                        count += 1
                        x += cell_width  # Move to the next column
                        continue
                
                # If normal orientation doesn't fit, try rotated (height x width)
                if x + cell_height <= panel_size and y + cell_width <= panel_size:
                    if np.all(grid[y:y+cell_width, x:x+cell_height] == 0):  # Ensure no overlap
                        cells.append((x, y, cell_height, cell_width))
                        grid[y:y+cell_width, x:x+cell_height] = 1
                        count += 1
                        x += cell_height  # Move to the next column
                        continue
                
                x += 1  # Move right if no cell fits
            else:
                x += 1  # Move right if space is occupied
        
        y += 1  # Move to the next row
    
    # Visualization using Matplotlib
    if show_plot:
        fig, ax = plt.subplots(figsize=(6, 6))
        ax.set_xlim(0, panel_size)
        ax.set_ylim(0, panel_size)
        ax.set_xticks(np.arange(0, panel_size + 1, cell_width))
        ax.set_yticks(np.arange(0, panel_size + 1, cell_height))
        ax.grid(True)

        # Draw cells
        for x, y, w, h in cells:
            rect = plt.Rectangle((x, y), w, h, edgecolor='blue', facecolor='lightblue', lw=2)
            ax.add_patch(rect)

        # Draw keep-out area
        if keep_out:
            kx, ky, kw, kh = keep_out
            ko_rect = plt.Rectangle((kx, ky), kw, kh, edgecolor='red', facecolor='red', alpha=0.5, lw=2)
            ax.add_patch(ko_rect)

        ax.set_title(f"Panel {panel_size}x{panel_size}: {count} Cells ({cell_width}x{cell_height})")
        plt.gca().invert_yaxis()  # Invert y-axis for correct display
        plt.show()

    return count, (panel_size, cell_width, cell_height, keep_out)

# Example Usage with Keep-Out Areas
panels = [
    (200, 70, 40, (50, 50, 80, 80)),  # Panel 200x200, Cells 70x40, Keep-Out area (50,50,80,80)
    (200, 50, 50, (100, 100, 50, 50)), # Panel 200x200, Cells 50x50, Keep-Out area (100,100,50,50)
    (150, 30, 60, None)  # No Keep-Out area
]

panel_results = []  # Store results for summary

for panel_size, cell_width, cell_height, keep_out in panels:
    num_cells, panel_info = fit_cells_in_panel(panel_size, cell_width, cell_height, keep_out)
    panel_results.append((panel_info, num_cells))

# Print Summary
print("\nSummary:")
print(f"Total Panels: {len(panel_results)}\n")
for i, (panel_info, num_cells) in enumerate(panel_results):
    panel_size, cell_width, cell_height, keep_out = panel_info
    ko_text = f"Keep-Out at {keep_out}" if keep_out else "No Keep-Out"
    print(f"Panel {i+1}: {panel_size}x{panel_size} - {num_cells} cells of size {cell_width}x{cell_height} ({ko_text})")
```
